```python
import pandas as pd
```

## Load Data

```python
df_normal = pd.read_csv('Normal.csv')
df_classification = pd.read_csv('Classification.csv')
df_time_series = pd.read_csv('TimeSeries.csv')

df_normal.head()
```

```python
df_classification.head()
```

```python
df_time_series.head()
```

## Analysis

```python
summary = pd.DataFrame({
    'Scenario': ['Normal', 'Classification', 'Time Series'],

    'CPU Mean (%)': [
        df_normal['cpu_usage'].mean(),
        df_classification['cpu_usage'].mean(),
        df_time_series['cpu_usage'].mean()
    ],

    'Memory Mean (%)': [
        df_normal['memory_usage'].mean(),
        df_classification['memory_usage'].mean(),
        df_time_series['memory_usage'].mean()
    ],

    'Temperature Mean (°C)': [
        df_normal['temperature'].mean(),
        df_classification['temperature'].mean(),
        df_time_series['temperature'].mean()
    ]
})

print(summary.round(2))
```
