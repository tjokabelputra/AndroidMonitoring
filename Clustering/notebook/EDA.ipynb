## Correlation Graph

```python
df = df.drop(columns=['battery_charging'])
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='Blues')
plt.title("Korelasi antar Fitur")
plt.show()
```

```python
df.hist(figsize=(8, 8), bins=20, edgecolor='black')
plt.suptitle("Distribusi Data Numerik")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.suptitle("Pairplot Fitur Numerik")
plt.show()
```

```python
X = df.select_dtypes(include=['float64','int64']).to_numpy()
model = KMeans(random_state=42)

visualizer = KElbowVisualizer(
    model,
    k=(1, 10),
    force_model=True
)

visualizer.fit(X)
best_k = visualizer.elbow_value_
print("Best K: ", best_k)
```

```python
plt.figure(figsize=(8,5))

plt.plot(visualizer.k_values_, visualizer.k_scores_, marker='o')

# Highlight elbow
plt.axvline(x=best_k, color='red', linestyle='--', label=f'Elbow (k={best_k})')

plt.title('Elbow Method')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('WCSS')
plt.legend()

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()

print("Best K:", best_k)
```
