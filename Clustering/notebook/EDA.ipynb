# Exploration Data Analysis

## Import Library

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv('../data/dataset/dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

## Graph Monitoring

### Hardware Monitoring

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["cpu_usage"], color="blue", linewidth=2)
plt.title("CPU Usage (Percent)")
plt.grid(axis="x", visible=False)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["memory_usage"], color="blue", linewidth=2)
plt.title("Memory Usage (Percent)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["temperature"], color="blue", linewidth=2)
plt.title("Temperature (°C)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

### Battery Monitoring

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["battery_level"], color="blue", linewidth=2)
plt.title("Battery Level (Percent)")
plt.grid(axis="x", visible=False)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["battery_drain"], color="blue", linewidth=2)
plt.title("Battery Drain (mA)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["battery_charging"], color="blue", linewidth=2)
plt.title("Battery Charging State (Charging/Not Charging)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

### Network Monitoring

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["upload_speed_kbps"], color="blue", linewidth=2)
plt.title("Upload Speed (kbps)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["download_speed_kbps"], color="blue", linewidth=2)
plt.title("Download Speed (kbps)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

### Screen State

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["screen_on"], color="blue", linewidth=2)
plt.title("Screen State (On/Off)")
plt.grid(axis="x", visible=False)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["brightness"], color="blue", linewidth=2)
plt.title("Screen Brightness")
plt.grid(axis="x", visible=False)
plt.ylim(0, 255)
plt.tight_layout()
plt.show()
```

## Correlation Graph

```python
df = df.drop(columns=['battery_charging'])
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f', cmap='Blues')
plt.title("Korelasi antar Fitur")
plt.show()
```

```python
df.hist(figsize=(8, 8), bins=20, edgecolor='black')
plt.suptitle("Distribusi Data Numerik")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df.select_dtypes(include=np.number))
plt.suptitle("Pairplot Fitur Numerik")
plt.show()
```
